```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# Define domain
dx = 1e-5
L = np.pi
x = L*np.arange(-1+dx,1+dx,dx)
n = len(x)
```

```python
# Define f function
quart = int(n/4)
f = np.sin(x)
f += 2*np.cos(5*x)
f += -2*np.sin(11*x)


FIG = plt.figure(figsize = (10,5))
colors = iter(plt.cm.rainbow(np.linspace(0, 0.8, 4)))

A0 = np.sum(f * np.ones_like(x)) * dx
fFS = A0/2

A_list = list()
B_list = list()

for k in tqdm(range(1, 14)):
    A = np.sum(f * np.cos(np.pi*(k)*x/L)) * dx # Inner product
    A_list.append(A)
    B = np.sum(f * np.sin(np.pi*(k)*x/L)) * dx
    B_list.append(B)
    fFS = fFS + A*np.cos((k)*np.pi*x/L) + B*np.sin((k)*np.pi*x/L)
    if k%4 == 1:
        plt.plot(x,fFS,'-', label = k, color = next(colors), linewidth = 7, alpha = 0.3)
plt.plot(x, f, label = 'original', linewidth = 1, color = 'red', alpha = 1)

plt.legend(loc = 'upper right')

FIG.savefig('fs1.jpeg')
```

```python
plt.plot(A_list), plt.plot(B_list)
```

```python

# Define f function
quart = int(n/4)
f = np.zeros_like(x)
f[:quart] = np.linspace(-20, 3, quart)
f[quart:2*quart] = 3
f[2*quart:3*quart] = -np.linspace(0, 3, quart)*np.linspace(0, 3, quart)*np.linspace(0, 3, quart) + 3
f[3*quart:] = -np.exp(np.linspace(3, 0, quart)) - 4
f = f-2



plt.figure(figsize = (10,5))
colors = iter(plt.cm.rainbow(np.linspace(0, 0.9, 10)))

A0 = np.sum(f * np.ones_like(x)) * dx
fFS = A0/2

A_list = list()
B_list = list()

for k in tqdm(range(1, 41)):
    A = np.sum(f * np.cos(np.pi*(k)*x/L)) * dx # Inner product
    A_list.append(A)
    B = np.sum(f * np.sin(np.pi*(k)*x/L)) * dx
    B_list.append(B)
    fFS = fFS + A*np.cos((k)*np.pi*x/L) + B*np.sin((k)*np.pi*x/L)
    if k%5 == 1:
        plt.plot(x,fFS,'-', label = k, color = next(colors), linewidth = 2, alpha = 0.5)
plt.plot(x, f, label = 'original', linewidth = 1, color = 'red')

plt.legend(loc = 'upper right')
FIG.savefig('fs2.jpeg')
```
