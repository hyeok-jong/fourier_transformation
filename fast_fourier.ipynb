```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 5]

from tqdm import tqdm

# Define domain
dx = 1e-3
L = np.pi
x = L*np.arange(-1+dx,1+dx,dx)
n = len(x)
```

```python
# Define f function
original = 0
original += 10*np.cos(11*x)
original += -21*np.cos(21*x)
original += 31*np.sin(31*x)
original += -47*np.sin(41*x)
original[:500] += np.linspace(0, 400, 500)
original[500:1000] -= np.linspace(-400, 0, 500)
original[1000:1500] += np.linspace(0, 600, 500)
original[1500:] += 600
noise =  original + 30*np.random.normal(size = original.shape)


print(len(original))
plt.plot(noise)
plt.plot(original)
```

```python
fft_output = np.fft.fft(noise)
len(fft_output)
```

```python
psd = fft_output * np.conj(fft_output) / len(x)/len(x)
```

```python
cancled = fft_output * (psd >1)
plt.figure(figsize = (15,3))
plt.plot(noise, label = 'noise', alpha = 0.5)
plt.plot(np.fft.ifft(cancled), label = 'cancled', alpha = 0.7)
plt.plot(original, label = 'original', alpha = 0.9)
plt.legend()
```
